```python
from environ import set_visible_devices
set_visible_devices('CPU')

from p2p.train import do_train
from p2p.agents import P2PAgent

def train_agents(agent_class, num_clients, st_cl, directed=False, agent_pars=None):
    do_train(agent_class,
         num_clients=num_clients,
         batch_size=50,
         model_pars={"model_v": 4, "lr": 0.005, "default_weights": True},
         agent_pars=agent_pars,
         graph_pars={'graph_type': 'sparse', 'num_neighbors': 3, 'directed': directed, 'time_varying': -1},
         epochs=100, seed=123, starting_client=st_cl)
```

# Effects of early stopping
## Directed communication

## Training without early stopping
### Training 100 agents

```python
train_agents(P2PAgent, 100, 0, directed=True, agent_pars={'early_stopping': False})
```

```python
train_agents(P2PAgent, 100, 100, directed=True, agent_pars={'early_stopping': False})
```

```python
train_agents(P2PAgent, 100, 200, directed=True, agent_pars={'early_stopping': False})
```

### Training 300 agents

```python
train_agents(P2PAgent, 300, 0, directed=True, agent_pars={'early_stopping': False})
```

```python
train_agents(P2PAgent, 300, 300, directed=True, agent_pars={'early_stopping': False})
```

```python
train_agents(P2PAgent, 300, 600, directed=True, agent_pars={'early_stopping': False})
```
